# Trabalhando com a base de convites

```python
import pandas as pd
convites = pd.read_csv('/content/drive/MyDrive/IA EXPERT ACADEMY/02. Mineração e Análise de Dados do LinkedIn/Mineração e Análise de Dados do LinkedIn/convites_localizacao.csv')
convites.drop(labels= ['Unnamed: 0'], axis=1, inplace=True)
convites.head()
```

Essa base de dados armazena o nome e ifnormações de localização das **empresas** das pessoas que mandaram solicitações pro detentor dessa base de dados

```python
import plotly.express as px
```

```python
grafico = px.treemap(convites[convites['Company'].notnull()], path=['Company'])
grafico.show()
```

```python
# Vamos fazer um filtro para convites e tirar os nulls
convites2 = convites.loc[(convites['Pais'].notnull()) & (convites['Company'].notnull())]
```

```python
convites2.isnull().sum()
```

```python
grafico = px.treemap(convites2, path=['Pais', 'Cidade', 'Company'])
grafico.show()
```
