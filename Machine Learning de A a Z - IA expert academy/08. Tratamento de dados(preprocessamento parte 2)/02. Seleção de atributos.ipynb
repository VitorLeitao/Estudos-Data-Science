```python
import pandas as pd
base_census = pd.read_csv('/content/drive/MyDrive/Machine Learning e Data Science com Python de A a Z/Bases de dados/census.csv')
base_census.head()
```

```python
# Criamops um labelEncoder para cada um dos atributos
from sklearn.preprocessing import LabelEncoder
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()
# aplicamos um labelEncoder para cada um dos atributos
X_census[:,1] = label_encoder_workclass.fit_transform(X_census[:,1])
X_census[:,3] = label_encoder_education.fit_transform(X_census[:,3])
X_census[:,5] = label_encoder_marital.fit_transform(X_census[:,5])
X_census[:,6] = label_encoder_occupation.fit_transform(X_census[:,6])
X_census[:,7] = label_encoder_relationship.fit_transform(X_census[:,7])
X_census[:,8] = label_encoder_race.fit_transform(X_census[:,8])
X_census[:,9] = label_encoder_sex.fit_transform(X_census[:,9])
X_census[:,13] = label_encoder_country.fit_transform(X_census[:,13])
```

```python
X_census
```

```python
# Normalização dos dados
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_census_scaled = scaler.fit_transform(X_census)
X_census_scaled
```

# Low Variance

Vão ser escolhidos os melhores atributos com base na variancia, buscando aqueles atributos que possuem uma variancia MAIOR, uma vez que geralmente aqueles que tem variancia muito baixa, sao parecidos, enquanto os maiores sao bastante diferentes, ajudando no treinamento do modelo

```python
import numpy as np
```

```python
X_census.shape[1] # Indica a quantidade de COLUNAS
```

```python
for i in range(X_census.shape[1]): # iterar sobre o numero de colunas
  print(X_census_scaled[:, i].var()) # Vai mostrar a variancia de cada coluna
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selecao = VarianceThreshold(threshold=0.05) # Variancia minima
X_census_variancia = selecao.fit_transform(X_census_scaled)
X_census_variancia.shape
```

```python
# PEgando os indices dos atributos selecionados
index = np.where(selecao.variances_ > 0.05)
index
```

```python
colunas[index]
```

```python
# Vendo se os resultados melhoraram
base_census_variancia = base_census.drop(columns = ['age', 'workclass', 'final-weight',
                                                    'education-num', 'race', 'capital-gain',
                                                    'capital-loos', 'hour-per-week',
                                                    'native-country'], axis = 1)
base_census_variancia
```

Nova base de dados com as colunas aprovadas pelo low variance

```python
X_census_variancia = base_census_variancia.iloc[:, 0:5].values
y_census_variancia = base_census_variancia.iloc[:, 5].values
X_census_variancia.shape, y_census_variancia.shape
```

```python
X_census_variancia
```

```python
X_census_variancia[:,0] = label_encoder_education.fit_transform(X_census_variancia[:,0])
X_census_variancia[:,1] = label_encoder_marital.fit_transform(X_census_variancia[:,1])
X_census_variancia[:,2] = label_encoder_occupation.fit_transform(X_census_variancia[:,2])
X_census_variancia[:,3] = label_encoder_relationship.fit_transform(X_census_variancia[:,3])
X_census_variancia[:,4] = label_encoder_sex.fit_transform(X_census_variancia[:,4])
```

```python
X_census_variancia
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
onehotencorder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [0,1,2,3,4])],remainder='passthrough')
X_census_variancia = onehotencorder.fit_transform(X_census_variancia).toarray()
X_census_variancia
```

```python
# Normalização dos dados

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_census_scaled = scaler.fit_transform(X_census_variancia)
X_census_scaled
```

```python
# Dividindo a base de dados
from sklearn.model_selection import train_test_split
X_census_treinamento_var, X_census_teste_var, y_census_treinamento_var, y_census_teste_var = train_test_split(X_census_variancia, y_census_variancia, test_size=0.15, random_state=0)
X_census_treinamento_var.shape, X_census_teste_var.shape
```

```python
# 84.70 sem trabalhar com seleção de atributos
from sklearn.ensemble import RandomForestClassifier
random_forest_var = RandomForestClassifier(criterion = 'entropy', min_samples_leaf =  1, min_samples_split = 5, n_estimators = 100)
random_forest_var.fit(X_census_treinamento_var, y_census_treinamento_var)
```

```python
from sklearn.metrics import accuracy_score, classification_report
previsoes = random_forest_var.predict(X_census_teste_var)
accuracy_score(y_census_teste_var, previsoes)
```

- Piorou o resultado
- em geral esse tipo de estrategia funciona melhor quando trabalhamos com imagens ou com base de dados com MUITOS atributos

# Extra tree

Treinamremos e veremos quais foram os atributos mais importante para essa base

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
base_census = pd.read_csv('/content/drive/MyDrive/Machine Learning e Data Science com Python de A a Z/Bases de dados/census.csv')
X_census = base_census.iloc[:, 0:14].values
y_census = base_census.iloc[:, 14].values
from sklearn.preprocessing import LabelEncoder
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()
# aplicamos um labelEncoder para cada um dos atributos
X_census[:,1] = label_encoder_workclass.fit_transform(X_census[:,1])
X_census[:,3] = label_encoder_education.fit_transform(X_census[:,3])
X_census[:,5] = label_encoder_marital.fit_transform(X_census[:,5])
X_census[:,6] = label_encoder_occupation.fit_transform(X_census[:,6])
X_census[:,7] = label_encoder_relationship.fit_transform(X_census[:,7])
X_census[:,8] = label_encoder_race.fit_transform(X_census[:,8])
X_census[:,9] = label_encoder_sex.fit_transform(X_census[:,9])
X_census[:,13] = label_encoder_country.fit_transform(X_census[:,13])
scaler = MinMaxScaler()
X_census_scaled = scaler.fit_transform(X_census)
```

```python
X_census_scaled.shape
```

```python
selecao = ExtraTreesClassifier()
selecao.fit(X_census_scaled, y_census)
```

```python
importancia = selecao.feature_importances_
importancia
```

```python
index = []
for i in range(len(importancia)):
  if importancia[i] >= 0.029:
    index.append(i)
```

```python
colunas[index]
```

```python
X_census_extra = X_census[:, index]
X_census_extra
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
onehotencorder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), [1,3,5,6,7])],remainder='passthrough')
X_census_extra = onehotencorder.fit_transform(X_census_extra).toarray()
X_census_extra
```

```python
X_census_extra.shape
```

```python
# Dividindo a base de dados
from sklearn.model_selection import train_test_split
X_census_treinamento_extra, X_census_teste_extra, y_census_treinamento_extra, y_census_teste_extra = train_test_split(X_census_extra, y_census, test_size=0.15, random_state=0)
X_census_treinamento_extra.shape, X_census_teste_extra.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_extra = RandomForestClassifier(criterion = 'entropy', min_samples_leaf =  1, min_samples_split = 5, n_estimators = 100)
random_forest_extra.fit(X_census_treinamento_extra, y_census_treinamento_extra)
```

```python
# 0.8470
from sklearn.metrics import accuracy_score, classification_report
previsoes = random_forest_extra.predict(X_census_teste_extra)
accuracy_score(y_census_teste_extra, previsoes)
```

Não mudou muita coisa...

Apesar de o accurancy nao ter mudado, a aplicação do onehot encoder resultou num numero bem menor de colunas, logo, o tempo de processamento é bem menor, resultando em um ganho, mesmo nao sendo em relação ao accurancy
